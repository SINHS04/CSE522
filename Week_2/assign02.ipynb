```python
import matplotlib.pyplot as plt
import numpy as np
```

# 실습 1

```python
def cal_distance(x, y):
  return np.sqrt(np.power(x[0] - y[0], 2) + np.power(x[1] - y[1], 2))

x = np.array([-3, 2])
y = np.array([3, 2])
distance = cal_distance(x, y)

# plot 코드를 작성하시오
plt.arrow(0, 0, x[0], x[1], color = 'red', head_width = .2, head_length = .2, label = 'vector x')
plt.arrow(0, 0, y[0], y[1], color = 'blue', head_width = .2, head_length = .2, label = 'vector y')
plt.plot([x[0], y[0]], [x[1], y[1]], linestyle = '--', marker = 'o', color = 'green', label = '||x-y||')
plt.hlines(0, -4, 4, linestyle = '--', color = 'gray')
plt.vlines(0, -4, 4, linestyle = '--', color = 'gray')
plt.axis([-4, 4, -4, 4])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='upper right')
plt.grid(True)

plt.show()

print(f'두 벡터 사이의 거리는 {distance}')
```

# 실습 2

```python
def cal_cos_theta(x, y):
    # theta 구하는 코드..
    # dot_product = np.dot(x, y)
    # p_Norm = cal_distance(x, np.array([0, 0])) * cal_distance(y, np.array([0, 0]))
    # return np.arccos(dot_product / p_Norm)

    return np.dot(x / cal_distance(x, np.array([0, 0])), y / cal_distance(y, np.array([0, 0])))

x = np.array([-3, 2])
y = np.array([3, 2])
x_unit = x / cal_distance(x, np.array([0, 0]))
y_unit = y / cal_distance(y, np.array([0, 0]))

cos_theta_1 = cal_cos_theta(x, y)
cos_theta_2 = cal_cos_theta(x_unit, y_unit)

# plot 코드를 작성하시오
plt.arrow(0, 0, x[0], x[1], color = 'red', head_width = .2, head_length = .2, label = 'vector x')
plt.arrow(0, 0, y[0], y[1], color = 'blue', head_width = .2, head_length = .2, label = 'vector y')
plt.arrow(0, 0, x_unit[0], x_unit[1], color = 'pink', head_width = .2, head_length = .2, label = 'vector x_')
plt.arrow(0, 0, y_unit[0], y_unit[1], color = 'skyblue', head_width = .2, head_length = .2, label = 'vector y_')
plt.hlines(0, -4, 4, linestyle = '--', color = 'gray')
plt.vlines(0, -4, 4, linestyle = '--', color = 'gray')
plt.axis([-4, 4, -4, 4])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='lower left')
plt.grid(True)

plt.show()

print(f'x, y 사이의 cos theta 값은 {cos_theta_1}')
print(f'x_, y_ 사이의 cos theta 값은 {cos_theta_2}')
```

# 실습 3

```python
def projection(origin_v, target_v) :
    target_unit = target_v / cal_distance(target_v, np.array([0, 0]))
    return np.dot(origin_v, target_unit) * target_unit

origin_v = np.array([3, 4])
target_v = np.array([4, 0])
proj_v = projection(origin_v, target_v)

# plot 코드를 작성하시오
plt.arrow(0, 0, origin_v[0], origin_v[1], color = 'red', head_width = .2, head_length = .2, label = 'Original Vector')
plt.arrow(0, 0, target_v[0], target_v[1], color = 'blue', head_width = .2, head_length = .2, label = 'target vector')
plt.arrow(0, 0, proj_v[0], proj_v[1], color = 'green', head_width = .2, head_length = .2, label = 'Projected Vector')
plt.plot([origin_v[0], proj_v[0]], [origin_v[1], proj_v[1]], linestyle = '--', marker = 'o', color = 'yellow')
plt.hlines(0, -1, 7, linestyle = '--', color = 'gray')
plt.vlines(0, -1, 7, linestyle = '--', color = 'gray')
plt.axis([-1, 7, -1, 7])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='upper right')
plt.grid(True)

plt.show()

print(f'origin_vector:', origin_v)
print(f'target_vector:', target_v)
print(f'projected vector: ', proj_v)
```

# 소감 및 느낀점

이번 과제를 통해 pyplot을 사용해보았다. pyplot은 프로그래밍 언어가 아닌 마크업 언어처럼 느껴졌다. 논리적이기 보단 구조적이기 때문이었다.
물론 이번 과제가 비교적 간단해서 그런것일수도 있지만, 과제를 수행함에 있어 크게 어려움은 없었다.

이론시간에 배운 내용을 직접 실습해서 그런지 프로그래밍이 더 재미있었다. 수학적인 내용을 코드로 구현한것은 오랜만이라 코드를 구현할 때 푹 빠져있었다.
Vector Projection도 직접 구현해보니 이해가 더 잘되었고, 더 기억에 남을 것 같다.
이것 뿐만이 아니라 단위벡터 구하기, 라이브러리를 사용한 내적 등 직접 코딩을 해봤기 때문에 기억에 오래 남을 것 같다.

앞으로 많은 것을 배울 텐데 그 내용을 코드로 구현하는것이 벌써 기다려진다.
